```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set(font_scale=2.5)
sns.set(style="whitegrid")
plt.style.use('seaborn')

import warnings
warnings.filterwarnings('ignore')
```

```python
diamonds = sns.load_dataset('diamonds')
diamonds.shape
```

```python
diamonds.head()
```

```python
diamonds.describe()
```

```python
# Shift + Tab      .
sns.scatterplot(data=diamonds, x='carat', y="price")
```

```python
plt.figure(figsize=(6.5, 6.5))
sns.scatterplot(data=diamonds, x='carat', y="price")
```

```python
plt.figure(figsize=(6.5, 6.5))
sns.scatterplot(data=diamonds, x='carat', y="price", hue="clarity")
```

```python
plt.figure(figsize=(6.5, 6.5))
clarity_ranking = ["I1", "SI2", "SI1", "VS2", "VS1", "VVS2", "VVS1", "IF"]
sns.scatterplot(data=diamonds, x='carat', y="price", hue="clarity", hue_order=clarity_ranking)
```

```python
plt.figure(figsize=(6.5, 6.5))
clarity_ranking = ["I1", "SI2", "SI1", "VS2", "VS1", "VVS2", "VVS1", "IF"]
sns.scatterplot(data=diamonds, x='carat', y="price", hue="clarity", hue_order=clarity_ranking, size="depth")
```

```python
plt.figure(figsize=(6.5, 6.5))
clarity_ranking = ["I1", "SI2", "SI1", "VS2", "VS1", "VVS2", "VVS1", "IF"]
sns.scatterplot(data=diamonds, x='carat', y="price", hue="clarity", hue_order=clarity_ranking, size="depth",
                palette="ch:r=-.2, d=.3_r")
```

```python
plt.figure(figsize=(6.5, 6.5))
clarity_ranking = ["I1", "SI2", "SI1", "VS2", "VS1", "VVS2", "VVS1", "IF"]
sns.scatterplot(data=diamonds, x='carat', y="price", hue="clarity", hue_order=clarity_ranking,
                size="depth", sizes=(1,8), linewidth=0,
                palette="ch:r=-.2, d=.3_r")
```

```python
plt.figure(figsize=(6.5, 6.5))
clarity_ranking = ["I1", "SI2", "SI1", "VS2", "VS1", "VVS2", "VVS1", "IF"]
sns.scatterplot(data=diamonds, x='carat', y="price", hue="clarity", hue_order=clarity_ranking,
                size="depth", sizes=(1,8), linewidth=0,
                palette="ch:r=-.2, d=.3_r")
sns.despine(left=True, bottom=True) #   .     .
```
