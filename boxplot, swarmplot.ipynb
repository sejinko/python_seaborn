```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
sns.set(style='ticks')
plt.style.use('seaborn')
```

```python
planets = sns.load_dataset('planets')
planets.head()
```

```python
planets.shape
```

```python
planets.info()
```

```python
sns.boxplot(data=planets, x='distance', y='method')
```

값의 차이가 너무 많이 나면 log를 씌워서 보정을 해주고, 차이가 거의 없는데 값의 미묘한 차이를 보고 싶다고 할때는 제곱(power transform)을 해서 값을 본다.

```python
planets['distance_log'] = np.log(planets['distance'])
```

```python
sns.boxplot(data=planets, x='distance_log', y='method')
```

swarmplot은 violineplot과 마찬가지로 산점도를 표현하게 된다.

```python
sns.boxplot(data=planets, x='distance_log', y='method')
sns.swarmplot(data=planets, x='distance_log', y='method')
```

```python
sns.boxplot(data=planets, x='distance_log', y='method')
sns.swarmplot(data=planets, x='distance_log', y='method', size=2, color=".3", linewidth=0)
```

ax.set_xscale('log')를 사용해서 log로 변환할 수 있다.

```python
f, ax = plt.subplots(figsize=(7,6))
ax.set_xscale('log')

sns.boxplot(data=planets, x='distance', y='method')
sns.swarmplot(data=planets, x='distance', y='method', size=2, color=".3", linewidth=0)
```

```python
f, ax = plt.subplots(figsize=(7,6))
ax.set_xscale('log')

sns.boxplot(data=planets, x='distance', y='method')
sns.swarmplot(data=planets, x='distance', y='method', size=2, color=".3", linewidth=0)

ax.xaxis.grid(True)
ax.set(ylabel="")
sns.despine(trim=True, left=True)
```
