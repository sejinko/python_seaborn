```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
sns.set(font_scale=1.5, style="white", palette='muted', color_codes=True)
```

```python
sns.color_palette("muted")
```

```python
sns.palplot(sns.color_palette("muted"))
```

```python
rs = np.random.RandomState(10)
d = rs.normal(size=100)
print(len(d))
print(d)
```

distplot은 Series, 1d-array, list를 표현한다.

```python
sns.distplot(d)
```

```python
sns.distplot(d, bins=30)
```

```python
sns.distplot(d, color='g', kde=False)
```

rug는 카펫의 실이다. rug를 True로 하면 값이 어디에 몰려 있는지를 볼 수 있다.

```python
sns.distplot(d, hist=False, rug=True, color='r')
```

```python
sns.distplot(d, color="g", hist=False, kde_kws={"shade":True})
```

```python
g = sns.distplot(d, color='m')
# 그래프를 변수에 할당하면 위에 글자가 나오지 않는다.
```

```python
f, ax = plt.subplots(2,2, figsize=(8,8), sharex=True)
#sharex를 True로 사용하면 x축은 하나로만 사용하겠다는 것이다.
sns.despine(left=True)
#despine은 테두리를 그리지 않겠다는 것이다.
```

```python
f, ax = plt.subplots(2,2, figsize=(8,8), sharex=True)
sns.despine(left=True)

sns.distplot(d, kde=False, ax=ax[0,0])
sns.distplot(d, hist=False, rug=True, color='r', ax=ax[0,1])

sns.distplot(d, color='m')
```

```python
f, ax = plt.subplots(2,2, figsize=(8,8), sharex=True)
sns.despine(left=True)

sns.distplot(d, kde=False, ax=ax[0,0])
sns.distplot(d, hist=False, rug=True, color='r', ax=ax[0,1])

sns.distplot(d, color='m')

plt.setp(ax, yticks=[])
plt.tight_layout()
```
