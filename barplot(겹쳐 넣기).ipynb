```python
import seaborn as sns
import matplotlib.pyplot as plt
sns.set(style='whitegrid')
```

```python
crashes = sns.load_dataset('car_crashes')
crashes.head()
```

```python
crashes.sample()
```

```python
crashes = crashes.sort_values('total', ascending=False)
crashes.head()
```

```python
sns.barplot(x='total', y='abbrev', data=crashes)
```

```python
plt.figure(figsize=(5,10))
sns.barplot(x='total', y='abbrev', data=crashes)
```

```python
plt.figure(figsize=(5,10))
sns.barplot(x='total', y='abbrev', data=crashes, color='b')
```

```python
plt.figure(figsize=(5,10))
sns.set_color_codes('pastel')
sns.barplot(x='total', y='abbrev', data=crashes, color='b')

sns.set_color_codes('muted')
sns.barplot(x='alcohol', y='abbrev', data=crashes, color='b')
```

```python
f, ax = plt.subplots(figsize=(5,10))

sns.set_color_codes('pastel')
sns.barplot(x='total', y='abbrev', data=crashes, color='b')

sns.set_color_codes('muted')
sns.barplot(x='alcohol', y='abbrev', data=crashes, color='b')

ax.legend(ncol=2, loc='lower right', frameon=True)
ax.set(xlim=(0, 24), ylabel='', xlabel='Automobile collisions per billion miles')
```
