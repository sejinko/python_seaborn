```python
import seaborn as sns
sns.set(style='white')
```

```python
iris_df = sns.load_dataset('iris')
```

```python
iris_df.head()
```

```python
sns.pairplot(iris_df, hue='species')
```

아래처럼 scatterplot을 입력하면 pairplot과 동일하게 나온다.

```python
g = sns.PairGrid(iris_df, hue='species', diag_sharey=False)
g.map(sns.scatterplot)
```

```python
g = sns.PairGrid(iris_df, hue='species', diag_sharey=False)

g.map_lower(sns.kdeplot)
g.map_upper(sns.scatterplot)
g.map_diag(sns.kdeplot, lw=3)
```

PairGrid는 다변수를 그래프화 하는데 자주 사용된다.
