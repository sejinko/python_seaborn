```python
import matplotlib.pyplot as plt
import seaborn as sns

sns.set(style='darkgrid')
iris = sns.load_dataset('iris')
```

```python
setosa = iris.query("species == 'setosa'")
virginica = iris.query("species == 'virginica'")
```

```python
setosa.head()
```

```python
f, ax = plt.subplots(1,1, figsize=(8,8))
sns.kdeplot(setosa['sepal_width'])
```

```python
f, ax = plt.subplots(1,1, figsize=(8,8))
sns.kdeplot(setosa['sepal_width'], setosa['sepal_length'])
```

```python
f, ax = plt.subplots(1,1, figsize=(8,8))
sns.kdeplot(setosa['sepal_width'], setosa['sepal_length'], shade=True, shade_lowest=False, cmap='Blues')
```

```python
f, ax = plt.subplots(1,1, figsize=(8,8))
sns.kdeplot(setosa['sepal_width'], setosa['sepal_length'], shade=True, shade_lowest=False, cmap='Blues')
sns.scatterplot(setosa['sepal_width'], setosa['sepal_length'], data=setosa)
```

```python
f, ax = plt.subplots(1,1, figsize=(8,8))
sns.kdeplot(setosa['sepal_width'], setosa['sepal_length'], shade=True, shade_lowest=False, cmap='Blues')
sns.scatterplot(setosa['sepal_width'], setosa['sepal_length'], data=setosa)
sns.kdeplot(virginica['sepal_width'], virginica['sepal_length'])
sns.scatterplot(virginica['sepal_width'], virginica['sepal_length'])
```

```python
f, ax = plt.subplots(1,1, figsize=(8,8))
ax.set_aspect()
ax[0] = sns.kdeplot(setosa['sepal_width'], setosa['sepal_length'], shade=True, shade_lowest=False, cmap='Blues')
sns.scatterplot(setosa['sepal_width'], setosa['sepal_length'], data=setosa)
ax[0] = sns.kdeplot(virginica['sepal_width'], virginica['sepal_length'])
sns.scatterplot(virginica['sepal_width'], virginica['sepal_length'])
```
