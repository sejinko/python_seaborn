```python
import seaborn as sns

sns.set(style='darkgrid')
%matplotlib inline
```

```python
import pandas as pd

tips = sns.load_dataset('tips')
print(tips.shape)
tips.head()
```

```python
distplot, heatmap 을 제외 하고는 대부분 x,y,data를 사용한다.
```

```python
sns.jointplot(x='total_bill', y='tip', data=tips)
```

```python
g = sns.jointplot(x='total_bill', y='tip', data=tips, kind='reg')
```

```python
tips[['total_bill', 'tip']].corr()
```

```python
g = sns.jointplot(x='total_bill', y='tip', data=tips, kind='reg', xlim=(0, 60), ylim=(0, 12), color="m", height=7)
```

```python
#residual(잔차)
g = sns.jointplot(x='total_bill', y='tip', data=tips, kind='resid')
```

```python
g = sns.jointplot(x='total_bill', y='tip', data=tips, kind='kde')
```

```python
g = sns.jointplot(x='total_bill', y='tip', data=tips, kind='hex')
```
