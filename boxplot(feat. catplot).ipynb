```python
import seaborn as sns
sns.set(style='ticks', palette='pastel')
```

```python
tips = sns.load_dataset('tips')
tips.head()
```

```python
tips.info()
```

```python
sns.boxplot(x='day', y='total_bill', data=tips)
```

```python
sns.boxplot(x='day', y='total_bill', data=tips, hue='smoker')
```

```python
sns.boxplot(x='day', y='total_bill', data=tips, hue='smoker', palette=['m', 'g'])
```

```python
sns.boxplot(x='day', y='total_bill', data=tips, hue='smoker', palette=['m', 'g'])
sns.despine(offset=10, trim=True)
```

```python
sns.boxplot(x='day', y='total_bill', data=tips, hue='time')
```

order를 사용하면 순서가 바뀐다. order를 쓰지 않을때는 Lunch, Dinner순으로 나오는데 밑에처럼 order를 사용하면 Dinner, Lunch로 나온다.

```python
sns.boxplot(x='time', y='tip', data=tips, order=['Dinner', 'Lunch'])
```

catplot의 요소에다가 kind='box'를 넣어주면 boxplot을 그리게 된다. 그리고 col에다가 time을 넣어주면 time은 Lunch, Dinner로 구성되어 있으므로 그 2개의 subplot을 그려주게 된다.

```python
sns.catplot(x='sex', y='total_bill', hue='smoker', col='time', data=tips, kind='box', height=4, aspect=.7)
```

```python
sns.catplot(x='sex', y='total_bill', hue='smoker', col='time', data=tips, kind='bar', height=4, aspect=.7)
```
