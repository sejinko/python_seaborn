왜 그래프를 그려봐야 하는가?
평균과 표준편차 상관관계가 같더라도 분포가 다르면 완전 데이터가 다르기 때문에 그래프를 그려보고 확인한다.
https://autodeskresearch.com/publications/samestats

```python
import seaborn as sns
import pandas as pd

sns.set(style='whitegrid')
```

```python
diamonds = sns.load_dataset('diamonds')
diamonds.head()
```

가끔 숫자데이터가 object로 되어 있는 경우가 있다. 그 경우 int나 float로 바꿔주고 진행한다.

```python
diamonds.info()
```

```python
sns.boxenplot(data=diamonds, x='clarity', y='carat')
```

```python
clarity_ranking = ['I1', 'SI2', 'SI1', 'VS2', 'VS1', 'VVS2', 'VVS1', 'IF']

sns.boxenplot(data=diamonds, x='clarity', y='carat', order=clarity_ranking)
```

scale을 사용하면 이상치 거리가 있는것을 가깝게 보이도록 그림을 조정한다.

```python
clarity_ranking = ['I1', 'SI2', 'SI1', 'VS2', 'VS1', 'VVS2', 'VVS1', 'IF']

sns.boxenplot(data=diamonds, x='clarity', y='carat', order=clarity_ranking, scale='linear')
```

```python
clarity_ranking = ['I1', 'SI2', 'SI1', 'VS2', 'VS1', 'VVS2', 'VVS1', 'IF']

sns.boxenplot(data=diamonds, x='clarity', y='carat', order=clarity_ranking, scale='linear', color='b')
```

```python
sns.boxplot(data=diamonds, x='clarity', y='carat')
```

```python
sns.boxplot(data=diamonds, x='clarity', y='carat', color='b', order=clarity_ranking)
```

boxenplot은 값들이 어디에 많이 몰려있는지까지 볼 수 있다.

```python
sns.violinplot(data=diamonds, x='clarity', y='carat')
```

violinplot은 distplot을 세로로 세운 그래프이다. 값이 어느 범위에 몰려있는지를 확인해 볼 수 있다.(값의 분포)
