```python
import numpy as np
import seaborn as sns
sns.set(style='whitegrid')
```

randomstate를 써서 랜덤 값이 똑같이 나오도록 고정을 해줬다.

```python
rs = np.random.RandomState(7)
x = rs.normal(2, 1, 75)
y = 2 + 1.5 * x + rs.normal(0, 2, 75)
```

```python
x
```

```python
sns. scatterplot(x=x, y=y)
```

```python
sns. scatterplot(x=x, y=y, color='g')
```

```python
sns.regplot(x=x, y=y, color='g')
```

lowess는 locally weighted robust scatterplot smoothing의 약자이다.
선형회귀로 했을떄 오차가 얼마나 나는지를 그래프로 그렸다고 보면 될 것이다.

```python
sns.regplot(x=x, y=y, color='g', lowess=True)
```

residual(잔여)의 값을 표시한다고 보면 된다.

```python
sns.residplot(x=x, y=y, color='g')
```

```python
sns.residplot(x=x, y=y, color='g', lowess=True)
```

상관계수가 어느정도인지를 시각화해서 보고자 할 때 그리면 좋을 것이다.
