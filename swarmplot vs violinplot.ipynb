```python
import numpy as np
import pandas as pd
import seaborn as sns
sns.set(style='whitgrid', palette='muted')
```

```python
iris = sns.load_dataset('iris')
iris.head()
```

```python
iris.info()
```

#### 카테고리컬한 데이터들이 산점도를 쓸 때 swarmplot을 쓴다고 생각하면 된다.

카테고리컬한 데이터를 scatterpot으로 표현하면 이쁘게 안나온다.

scatterplot은 수치형 데이터일때 사용한다.

melt는 column에 있는 값을 행으로 녹여낸다.(melt = 녹이다)

컬럼데이터에 있는 수치들을 value에다가 담았다라고 보면 된다.

```python
iris = pd.melt(iris, 'species', var_name = 'measurement')
iris.head()
```

```python
sns.scatterplot(x='measurement', y='value', data=iris)
```

```python
sns.scatterplot(x='measurement', y='value', hue='species' , data=iris)
```

```python
sns.swarmplot(x='measurement', y='value', hue='species', palette=['r','c','y'], data=iris)
```

```python
sns.swarmplot(x='measurement', y='value', palette=['r','c','y'], data=iris)
```

```python
tips = sns.load_dataset('tips')
tips.sample()
```

```python
sns.violinplot(x='day', y='total_bill', hue='smoker', split=True, data=tips)
```

```python
sns.violinplot(x='day', y='total_bill', hue='smoker', split=True, data=tips, inner='point')
```

violinplot을 산점도로 그린것이 swarmplot이다라고 보면 된다.

```python
sns.swarmplot(x='day', y='total_bill', hue='smoker', data=tips)
```

```python
sns.violinplot(x='day', y='total_bill', hue='smoker', split=True, data=tips)
sns.swarmplot(x='day', y='total_bill', hue='smoker', data=tips)
```
