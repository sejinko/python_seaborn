```python
import seaborn as sns
sns.set(style='whitegrid')
```

```python
titanic = sns.load_dataset('titanic')
titanic.shape
```

```python
titanic.head()
```

```python
sns.PairGrid(titanic, y_vars='survived', x_vars=['class','sex','who','alone'], height=5, aspect=.5)
```

```python
g = sns.PairGrid(titanic, y_vars='survived', x_vars=['class','sex','who','alone'])
g.map(sns.pointplot)
```

```python
g = sns.PairGrid(titanic, y_vars='survived', x_vars=['class','sex','who','alone','embark_town'],
                 height=5, aspect=.5)
g.map(sns.pointplot)
```

```python
g = sns.PairGrid(titanic, y_vars='survived', x_vars=['class','sex','who','alone','embark_town'],
                 height=5, aspect=.5)
g.map(sns.pointplot, scale=1.3, errwidth=4, color='xkcd:plum')
g.set(ylim=(0,1))
```

```python
g = sns.PairGrid(titanic, y_vars='survived', x_vars=['class','sex','who','alone','embark_town'],
                 height=5, aspect=.5)
g.map(sns.pointplot, scale=1.3, errwidth=4, color='xkcd:plum')
g.set(ylim=(0,1))
# despine   .
sns.despine(fig=g.fig, left=True)
```

```python
g = sns.PairGrid(titanic, y_vars='survived', x_vars=['class','sex','who','alone','embark_town'],
                 height=5, aspect=.5)
g.map(sns.barplot)
```

```python
g = sns.PairGrid(titanic, y_vars='survived', x_vars=['class','sex','who','alone','embark_town'],
                 height=5, aspect=.5)
g.map(sns.violinplot)
```
